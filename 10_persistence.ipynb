```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_google_genai import GoogleGenerativeAI
from dotenv import load_dotenv
from langgraph.checkpoint.memory import InMemorySaver
```

```python
load_dotenv()

llm = GoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.9)
```

```python
class JokeState(TypedDict):

    topic: str
    joke: str
    explanation: str
```

```python
def create_joke(state: JokeState):
    prompt = f"""generate a joke about {state['topic']}"""
    respone=llm.invoke(prompt)
    return {'joke':respone}

    
```

```python
def generate_explanation(state: JokeState):
    prompt = f"""explain the joke: {state['joke']}"""
    respone=llm.invoke(prompt)
    return {'explanation':respone}
```

```python
graph = StateGraph(JokeState)
graph.add_node('create_joke',create_joke)
graph.add_node('generate_explanation',generate_explanation)

graph.add_edge(START,'create_joke')
graph.add_edge('create_joke','generate_explanation')
graph.add_edge('generate_explanation',END)


checkpointer = InMemorySaver()

workflow = graph.compile(checkpointer=checkpointer)
```

```python
config1 = {"configurable":{"thread_id":"1"}}
workflow.invoke({'topic':'pizza'},config=config1)
```

```python
workflow.get_state(config1)
```

```python
list(workflow.get_state_history(config1))
```
